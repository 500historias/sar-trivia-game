```python
import pandas as pd
from io import StringIO
pd.set_option("display.max_columns", None)
#pd.set_option("display.max_rows", 100)
```

```python
# Define the custom sorting order
custom_order_column1 = {
    'Facil': 0,
    'Media': 1,
    'Dificil': 2
}
custom_order_column2 = {
    'Mitologia Griega': 0,
    'Biblia': 1,
    'Mitologia Latina': 2,
    'Autores latinos': 3,
    'Cultura general': 4
}

```

```python
df = pd.read_excel(r"C:/Users/chucp/Desktop/Work/BD Preguntas/Preguntas_de_la_trivia.xlsx")
df = df.dropna(how='all')
# Create a new column combining custom ordering for both columns
df['custom_order'] = df['Dificultad'].map(custom_order_column1).astype(str) + '_' + df['Categoria'].map(custom_order_column2).astype(str)
# Sort the DataFrame based on the new column
sorted_df = df.sort_values(by='custom_order')
#Drop the custom order column
sorted_df = sorted_df.drop('custom_order', axis=1)
sorted_df
```

```python
# Save the table values to a CSV file
sorted_df.to_csv('preguntas_ordenadas.csv', index=False)

# Save the table values to an Excel file
sorted_df.to_excel('preguntas_ordenadas_excel.xlsx', index=False)
```
